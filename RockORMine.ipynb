```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('Sonar_data.csv', header=None)
```

Separating X and Y

```python
X = dataset.drop(columns= 60, axis=1)
Y = dataset[60]
```

```python
print(X.shape)
print(Y.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, stratify= Y, random_state= 2)
```

Training the model

```python
classifier = svm.SVC(kernel= 'linear')
```

```python
classifier.fit(X_train, Y_train)
```

Model Evaluation -> How many times the model is predicting correctly

```python
# Accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
# Accuracy above 75% is good
print("Accuracy of training data:",round(training_data_accuracy, 2)*100)
```

```python
# Accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
# Accuracy above 75% is good
print("Accuracy on test data:", round(test_data_accuracy,2) * 100)
```
